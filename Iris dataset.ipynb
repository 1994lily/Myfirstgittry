### 3D Scatter plot

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns

# Load the iris dataset
iris = sns.load_dataset("iris")

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(iris['sepal_length'], iris['sepal_width'], iris['petal_length'])
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
ax.set_zlabel('Petal Length')
plt.show()
```

### Pair plot

```python
# Create a pair plot
sns.pairplot(iris, hue="species")
```

### Histogram

```python
# Create a histogram of the sepal length
sns.histplot(data=iris, x="sepal_length")
```

### Probability function

```python
# Compute the mean and standard deviation of sepal lengths for setosa flowers
setosa_sepal_length_mean = np.mean(iris.loc[iris['species'] == 'setosa', 'sepal_length'])
setosa_sepal_length_std = np.std(iris.loc[iris['species'] == 'setosa', 'sepal_length'])

# Define the Gaussian probability function
def gaussian(x, mu, sigma):
    return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)

# Compute the probability of an Iris flower with sepal length 5 belonging to the setosa species
x = 5
p_setosa_given_sepal_length_5 = gaussian(x, setosa_sepal_length_mean, setosa_sepal_length_std)

print(p_setosa_given_sepal_length_5)
```

### Univariate analysis

```python
# Compute the mean and standard deviation of sepal lengths
sepal_length_mean = np.mean(iris['sepal_length'])
sepal_length_std = np.std(iris['sepal_length'])

# Define the Gaussian probability function
def gaussian(x, mu, sigma):
    return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)

# Compute the probability density function for sepal lengths using the Gaussian distribution
x_values = np.linspace(3, 9, 100)
pdf_values = gaussian(x_values, sepal_length_mean, sepal_length_std)

# Plot the probability density function
sns.lineplot(x=x_values, y=pdf_values)
```

### Cumulative Distribution function

```python
# Compute the CDF for sepal lengths
sepal_length_cdf = np.cumsum(np.histogram(iris['sepal_length'], bins=10, density=True)[0])

# Plot the CDF
sns.lineplot(x=np.histogram(iris['sepal_length'], bins=10)[1][1:], y=sepal_length_cdf)
```

### Variance & standard deviation

```python
# Compute the variance and standard deviation of sepal lengths
sepal_length_var = np.var(iris['sepal_length'])
sepal_length_std = np.std(iris['sepal_length'])

# Print the results
print("Sepal length variance:", sepal_length_var)
print("Sepal length standard deviation:", sepal_length_std)
```

### Median, Percentiles & Quartiles

```python
# Compute the median, percentiles, and quartiles of sepal lengths
sepal_length_median = np.median(iris['sepal_length'])
sepal_length_percentiles = np.percentile(iris['sepal_length'], [25, 50, 75])
sepal_length_quartiles = np.quantile(iris['sepal_length'], [0.25, 0.5, 0.75])

# Print the results
print("Sepal length median:", sepal_length_median)
print("Sepal length percentiles:", sepal_length_percentiles)
print("Sepal length quartiles:", sepal_length_quartiles)
```

### Donut chart

```python
# Create a donut chart for the species variable
species_counts = iris['species'].value_counts()
fig, ax = plt.subplots()
ax.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%')
ax.add_artist(plt.Circle((0,0), 0.7, color='white'))

# Show the plot
plt.show()
```
