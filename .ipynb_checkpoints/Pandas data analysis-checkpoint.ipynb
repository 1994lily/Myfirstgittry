## Operations in Pandas

```python
import pandas as pd
import numpy as np

# Create a sample DataFrame with some missing values
data = {
    'name': ['John', 'Jane', 'Bob', 'Alice', 'Mike'],
    'age': [25,  np.nan,20, 35, 28],
    'salary': [50000, 60000, 45000, np.nan, 55000],
    'department': ['Sales', 'Marketing', 'IT', 'Sales', 'IT']
}
df = pd.DataFrame(data)
df
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.describe(include="O")
```

```python
df.describe(include="all")
```

```python
df.shape
```

```python
df.columns
```

```python
df.index
```

```python
# Check which values are null
print(df.isnull())
```

```python
# Fill the missing age values with the mean age
mean_age = df['age'].mean()
df['age'].fillna(mean_age, inplace=True)
df
```

```python
# Fill the missing salary values with the median salary
median_salary = df['salary'].median()
df['salary'].fillna(median_salary, inplace=True)
df
```

```python
# Group the DataFrame by department and compute the mean salary for each department
grouped_df = df.groupby('department').mean()
grouped_df
```

```python
# Create a pivot table showing the mean salary for each department by age group
age_bins = [0, 25, 30, 35, np.inf]
age_labels = ['<25', '25-30', '30-35', '35+']
df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels)
pivot_table = df.pivot_table(values='salary', index='department', columns='age_group', aggfunc='mean')
pivot_table
```

### Aggregating data

```python
# How many people of geneder male and female embarked at static C,S,Q.

## to perform our aggregation we need to use dataframe followed by groupby 
# provide the column or list of columns for which we want to perform aggregation. i.e. 'Sex'
# what value we want to want to aggregate. i.e 'PassengerId'
# provide a function or way in which we want to aggragate. i.e. count
## e.g mean, meadian, count, sum etc.

## Question : Find how many people of different genders are travlleing on titanic.
titanic_df.groupby('Sex')['PassengerId'].count()

## collecting the value of count of passenger of different gender.
# --> Aggregation column --"Sex"
# --> Aggregated column -- "PassengerId"
# --> function to aggragtate -- count()

# dataframe.groupby([Aggregation column])[Aggregation column].aggregation_function()
```

```python
## Question : Find out how many people survived and died in the titanic crash.
titanic_df.groupby('Survived')['PassengerId'].count()
```

```python
## --> Question
# Find how many people have boarderd titanic at different station.
 ## -- first is to select your dataframe and use groupby method
 ## --> to identify your aggreartion column and put in parathesis 
 ## --> is to specify the columns which I want to aggregate in square bracket. agggregated column
 ## --> aggragtion function  count()  

titanic_df.groupby('Embarked')['PassengerId'].count()
## Most of the people have boarderd from station S followed C and last Q.

# Find out what is the average fare at each embarking station.
    ## first step --> select your dataframe and use groupby method.
    ## second step --> aggregration column -- 'Embarked'
    ## third step --> aggregated column -- 'Fare'
    ## fourth step --> Aggregation function -- mean()
    

titanic_df.groupby('Embarked')['Fare'].mean()
## Average price is lowest for Station Q followed S and lastly C.

# Find out what is the total fare at each embarking station.
titanic_df.groupby('Embarked')['Fare'].sum()
## Most of the amount from total sales is from Station S , followed C and lastly Q.


# Find how many people are travelling in different class.
titanic_df.groupby('Pclass')['PassengerId'].count()

# Find what is the average age of different gender travellling on titanic.
titanic_df.groupby('Sex')['Age'].mean()

# Find out maximum and minumum age of differnt geneder travelling on titanic
titanic_df.groupby('Sex')['Age'].max()    
titanic_df.groupby('Sex')['Age'].min()  

## --> Question
## 1st --> where you have multiple agggregation column
## Find how many people of different gender survived the crash.
## scenario --> I have two aggregtion column 'Sex','Survived'  i.e. ['Sex','Survived']

    # 1st step -- > select dataframe and use groupby menthod
    # 2nd Step --> aggregation column ['Sex','Survived']
    # 3rd Step --> aggregated column 'PassengerId'
    # 4th Step --> aggragtion function count

titanic_df.groupby(['Sex','Survived'])['PassengerId'].count()

## What is the average price each gender paid for fare art differnt station.
titanic_df.groupby(['Sex','Embarked'])['Fare'].mean()

## what is the average price of fare for differnt gender traverlling in differnt classs and different station.
titanic_df.groupby(['Pclass','Sex','Embarked'])['Fare'].mean()

## 2nd --> where you have multiple aggregated column
# question : Find out what is the average fare and age of differnt gender travelling on tiatnic.

 ## 1st -- select your dataframe and use groupby
 ## 2nd -- choose your aggrergation column 'Sex' --> category
 ## 3rd -- choose tour aggregated column ['Age','Fare']  --> numerical
 ## 4th -- choose your aggregated function mean

titanic_df.groupby('Sex')[['Age','Fare']].mean()

# Add a new column family_size
titanic_df['family_size'] = titanic_df['SibSp'] + titanic_df['Parch'] +1

# question Find out what is the average fare and age and family size of differnt gender travelling on tiatnic.
titanic_df.groupby('Sex')[['Age','Fare','family_size']].mean()


## Question find out each gender and survived travelling in different class people 
#what is the average age , fare and family_size .
titanic_df.groupby(['Pclass','Sex','Survived'])[['Age','Fare','family_size']].mean()

## 3rd --> where you have multiple aggregation function --> .agg()

## Question find out each gender and survived travelling in different class people 
#what is the average, maximum, minimum age , fare and family_size .
import numpy as np


titanic_df.groupby(['Pclass','Sex','Survived'])[['Age','Fare','family_size']].agg([np.mean,np.max,np.min])
```
