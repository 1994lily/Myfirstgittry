## Operations in Pandas

```python
import pandas as pd
import numpy as np

# Create a sample DataFrame with some missing values
data = {
    'name': ['John', 'Jane', 'Bob', 'Alice', 'Mike'],
    'age': [25,  np.nan,20, 35, 28],
    'salary': [50000, 60000, 45000, np.nan, 55000],
    'department': ['Sales', 'Marketing', 'IT', 'Sales', 'IT']
}
df = pd.DataFrame(data)
df
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.describe(include="O")
```

```python
df.describe(include="all")
```

```python
df.shape
```

```python
df.columns
```

```python
df.index
```

```python
# Check which values are null
print(df.isnull())
```

```python
# Fill the missing age values with the mean age
mean_age = df['age'].mean()
df['age'].fillna(mean_age, inplace=True)
df
```

```python
# Fill the missing salary values with the median salary
median_salary = df['salary'].median()
df['salary'].fillna(median_salary, inplace=True)
df
```

```python
# Group the DataFrame by department and compute the mean salary for each department
grouped_df = df.groupby('department').mean()
grouped_df
```

```python
# Create a pivot table showing the mean salary for each department by age group
age_bins = [0, 25, 30, 35, np.inf]
age_labels = ['<25', '25-30', '30-35', '35+']
df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels)
pivot_table = df.pivot_table(values='salary', index='department', columns='age_group', aggfunc='mean')
pivot_table
```

### Filtering data

```python
#filter the dataset to include only passengers with age greater than or equal to 18
adult_passengers = titanic_df[titanic_df['Age'] >= 18]
adult_passengers
```

```python
#filter the dataset to include only passengers in first class
first_class_passengers = titanic_df[titanic_df['Pclass'] == 1]
first_class_passengers
```

### Adding and removing columns from dataset

```python
#Add a new column to the dataset
titanic_df['FamilySize'] = titanic_df['SibSp']+titanic_df['Parch']+1

titanic_df.head()
```

```python
#Remove a column from the dataset
titanic_df = titanic_df.drop('Cabin',axis=1)
titanic_df.head()
```

### Aggregating data

```python
# Group the dataset by passenger class and calculate the mean fare for each class
mean_fare_by_class = titanic_df.groupby('Pclass')['Fare'].mean()
mean_fare_by_class
```

```python
# Calculate the mean age of all passengers
mean_age = titanic_df['Age'].mean()

# Calculate the total number of passengers in the dataset
total_passengers = titanic_df['PassengerId'].count()

# Calculate the total fare paid by all passengers
total_fare = titanic_df['Fare'].sum()

# Calculate the minimum and maximum ages of all passengers
min_age = titanic_df['Age'].min()
max_age = titanic_df['Age'].max()

# Print the results
print('Mean age:', mean_age)
print('Total passengers:', total_passengers)
print('Total fare:', total_fare)
print('Minimum age:', min_age)
print('Maximum age:', max_age)
```

### Merging and Joining data

```python
df1 = titanic_df.loc[:5,('PassengerId','Name','Age','Sex')]
df2 = titanic_df.loc[:5,('PassengerId','Ticket','Fare')]
```

```python
df1.head()
```

```python
 df2.head()
```

```python
# Merge the two datasets on the 'PassengerId' column
merged_df = pd.merge(df1, df2, on='PassengerId')
merged_df
```

```python
# Concatenate two datasets along the rows
concatenated_rows = pd.concat([df1, df2])
concatenated_rows
```

```python
# Concatenate two datasets along the columns
concatenated_cols = pd.concat([df1, df2], axis=1)
concatenated_cols
```

```python
# Join two datasets on the 'PassengerId' column
joined_df = df1.set_index('PassengerId').join(df2.set_index('PassengerId'))
joined_df
```
